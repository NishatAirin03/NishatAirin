```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#read in the data

df = pd.read_csv(r'C:\Users\nsimr\Downloads\movies.csv')
```

```python
df.head()
```

```python
#looking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#datatypes for columns
df.dtypes
```

```python
#changing datatypes of columns
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)

df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 

```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(str)
```

```python
df.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
df.head()
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel('budget in film')
plt.ylabel('gross earnings')
plt.show()
```

```python
df.head()
```

```python
sns.regplot(x="budget", y="gross", data=df,scatter_kws={"color":"black"})
#sns.regplot(x="budget", y="gross", data=df,
                 #scatter_kws={"color": "black"}, line_kws={"color": "red"})
```

```python
#correlation of the movie features
df.corr(numeric_only=True,method='pearson')
```

```python
correlation_matrix=df.corr(numeric_only=True,method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title("Correlation matric for numeric features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
df_numerized=df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes

df_numerized.head()
        
```

```python
df.head()
```

```python
correlation_matrix=df_numerized.corr(numeric_only=True,method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title("Correlation matric for numeric features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```
